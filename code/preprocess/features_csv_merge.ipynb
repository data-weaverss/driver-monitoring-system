```python
data_path = '/content/drive/MyDrive/DMS/data/EDA'
mar_12_path = 'mar_ear_final/final_df_12_mar_ear_interpolated.csv'
mar_24_path = 'mar_ear_final/final_df_24_mar_ear_interpolated.csv'
yaw_12_path = 'yaw_pitch_final/final_df_12_yaw_pitch_interpolated.csv'
yaw_24_path = 'yaw_pitch_final/final_df_24_yaw_pitch_interpolated.csv'
phone_12_path = 'phone_detected_final/phonecall_drop_raw_12.xlsx'
phone_24_path = 'phone_detected_final/phonecall_drop_raw_24.xlsx'
```

```python
import pandas as pd
import os
```

```python
pd.set_option('display.max_colwidth', None)
```

# timestep 12

```python
mar_12 = pd.read_csv(os.path.join(data_path, mar_12_path))
yaw_12 = pd.read_csv(os.path.join(data_path, yaw_12_path))
phone_12 = pd.read_excel(os.path.join(data_path, phone_12_path))
```

```python
print(f"mar_12 length: {len(mar_12)}")
print(f"yaw_12 length: {len(yaw_12)}")
print(f"phone_12 length: {len(phone_12)}")
```

```python
phone_12.isna().sum()
```

```python
phone_12[phone_12['Phone_Detected'].isna()]['clip_id']
```

```python
phone_12[(phone_12['groups_12'] == 8) &
         (phone_12['clip_id'] == 'gZ-31(File25)_s4_face_2019-04-03-09;04;29_31_7.avi')]
```

```python
phone_12 = phone_12[~phone_12['Phone_Detected'].isna()]
```

```python
phone_12.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
phone_12.columns = [col if col != 'Phone_Detected' else 'phone_detected' for col in phone_12.columns]
```

```python
print(f"mar_12 length: {len(mar_12)}")
print(f"yaw_12 length: {len(yaw_12)}")
print(f"phone_12 length: {len(phone_12)}")
```

```python
merged_df = mar_12.merge(yaw_12[['frame_path', 'yaw', 'pitch', 'drop']].drop_duplicates(), on='frame_path', how='left')
```

```python
merged_df = merged_df.merge(phone_12[['frame_path', 'phone_detected', 'drop_col']].drop_duplicates(), on='frame_path', how='left')
```

```python
merged_df.columns
```

```python
merged_df['is_mar_ear_streak5plus'].unique()
```

```python
merged_df['drop'].unique()
```

```python
merged_df['drop_col'].unique()
```

```python
merged_df = merged_df[
    (merged_df['is_mar_ear_streak5plus'] == False) &
    (merged_df['drop'].isna()) &
    (merged_df['drop_col'].isna())
]
```

```python
merged_df.columns = [col.lower() for col in merged_df.columns]
```

```python
merged_df.columns
```

```python
merged_df = merged_df[['frame_path', 'label', 'camera', 'clip_id', 'frame_no', 'dataset', 'groups_12', 'mar', 'ear', 'yaw', 'pitch', 'phone_detected']]
```

```python
# 그룹별 개수 계산
group_counts = merged_df.groupby(['label', 'clip_id', 'groups_12']).size().reset_index(name='size')

# size가 12가 아닌 그룹만 필터링
not_12 = group_counts[group_counts['size'] != 12]

print(not_12)
```

```python
merged_df.to_csv('/content/drive/MyDrive/DMS/data/merged_df_12.csv', index=False)
```

```python
len(merged_df[['label', 'clip_id', 'groups_12']].drop_duplicates())
```

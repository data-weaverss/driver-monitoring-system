```python
import os
from ultralytics import YOLO
import torch
```

```python
# 경로 설정
dataset_path = '/content/drive/MyDrive/phone_finetunning_0415'
data_yaml_path = os.path.join(dataset_path, "data.yaml")
model_path = os.path.join(dataset_path, "best_2.pt")

# 모델 로드
model = YOLO(model_path)

# 파인튜닝 실행
model.train(
    data=data_yaml_path,
    epochs=40,
    batch=32,
    imgsz=224,
    single_cls=True,
    project='/content/drive/MyDrive/phone_finetunning_0415',
    name='exp1',  # 결과 저장 폴더 이름
    device="cuda" if torch.cuda.is_available() else "cpu",
    workers=8
)

print("✅ 모델 학습 완료! 결과는 project 경로에서 확인하세요.")
```

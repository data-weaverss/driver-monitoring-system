```python
import os
import shutil
```

## phonecall 만 2초 부근으로 추가

```python
# 경로 설정
root_dir = '/content/drive/MyDrive/DMS/data/DMD/dmd_frames_resized/PhoneCall'           # 세부 폴더들이 들어 있는 상위 폴더 경로
output_dir = '/content/drive/MyDrive/DMS/phone_finetunning_0415/data/dmd_phonecall'            # 복사된 이미지 저장할 경로

# 출력 폴더가 없다면 생성
os.makedirs(output_dir, exist_ok=True)

# 모든 하위 폴더 탐색
for subdir, dirs, files in os.walk(root_dir):
    for file in files:
        if file == 'frame000012.jpg':
            src_path = os.path.join(subdir, file)
            # 저장 파일 이름에 폴더 이름 붙이기
            folder_name = os.path.basename(subdir)
            dst_filename = f'{folder_name}_frame000012.jpg'
            dst_path = os.path.join(output_dir, dst_filename)

            shutil.copy2(src_path, dst_path)
            print(f"Copied: {src_path} → {dst_path}")

        elif file == 'frame000013.jpg':
            src_path = os.path.join(subdir, file)
            # 저장 파일 이름에 폴더 이름 붙이기
            folder_name = os.path.basename(subdir)
            dst_filename = f'{folder_name}_frame000013.jpg'
            dst_path = os.path.join(output_dir, dst_filename)

            shutil.copy2(src_path, dst_path)
            print(f"Copied: {src_path} → {dst_path}")
        else :
            continue
```
